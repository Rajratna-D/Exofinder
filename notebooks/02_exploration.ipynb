# Phase 2: Exploratory Visualization
This notebook visualizes the raw, unprocessed light curves of TESS targets from three classes:
1. **Confirmed Planets** (stars hosting verified transiting planets)
2. **Eclipsing Binaries** (binary stars that eclipse each other, mimic transits)
3. **False Positives** (instrumental noise, stellar activity, or background eclipses)

For each class, we plot 4 raw light curves to inspect the data quality and see if they are visually distinguishable before applying detrending or period searches.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Ensure plots look modern and clean
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
print("Setup complete. Ready to visualize raw light curves.")
```

## Class: Confirmed Planet
Below are the raw light curves for four targets classified as `confirmed_planet`.

```python
# Visualize raw light curve for TIC 231663901
file_path = '../data/raw/TIC_231663901.csv'
df = pd.read_csv(file_path)

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(df['time'], df['flux'], color='#1f77b4', alpha=0.6, label='Raw Flux')
ax.set_title("TIC 231663901 (Confirmed Planet) - Raw Light Curve")
ax.set_xlabel('Time (BTJD)')
ax.set_ylabel('Flux')
ax.grid(True, alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
```

## Summary of Visual Observations

At this raw, unprocessed stage:
- **Confirmed Planets**: The raw light curves are dominated by long-term stellar activity, instrumental ramps, and systematic noise. The individual transit dips (which are typically less than 1% or 10,000 ppm of the total flux) are **completely invisible** to the naked eye. The overall flux curve shows strong trends, curves, or steep slopes (instrumental settling).
- **Eclipsing Binaries**: Some of these show **very prominent, deep, V-shaped or U-shaped dips** (often 5% to 50% drop in flux) that are easily visible even in raw data, because the eclipsing companion is stellar-sized and blocks a huge portion of light. However, others (especially smaller EBs) are masked by stellar variability or have complex noise patterns.
- **False Positives**: These curves display a wide range of behaviors. Some show massive instrumental ramps (rapidly decreasing/increasing flux), others show high-frequency noise or periodic stellar pulsations, and some show no obvious dips at all.

### Conclusion
We **cannot** reliably distinguish confirmed planets from false positives or smaller eclipsing binaries by eye in the raw data. The planet transits are hidden beneath systematic trends and stellar variability. This highlights the absolute necessity of **Phase 3 (Detrending)** to flatten the curves and **Phase 4 (Period Search & Phase Folding)** to stack the signal and reveal the transit shape.
